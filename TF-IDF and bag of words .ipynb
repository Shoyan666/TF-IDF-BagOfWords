```python
# function to calculate how many times do we face particular word Terf Frequency(TF)
def tf(sentence,word):
    x = len(sentence.split()) # words in sentence
    a = len([i for i in re.split('[^a-z]', sentence) if i == word])
    return a/x # returning number of occurrenses
    
```

```python
# function which calculate all the certain values in other words all non-zero values 
idf = [] # empty list to assign here idf values
def counter(column): 
    counter1 = 0
    for i in list(column):
        if i > 0:
            counter1+=1
    return counter1
# starting the loop to check zero values
for i in range(1, tf_of_bow.shape[1]):
    y = counter(tf_of_bow.iloc[:, i])
    if y == 0:
        idf.append(y)
        continue
        # we can't divide any number on 0 so we have to avoid this error
    idf.append(np.log10(tf_of_bow.shape[1]/y)) 
idf
```

#### In cell above, I tried to represent my tf as a data frame but I faced some troubles and decided to do it in another way by using a function with the list and then representing this list as the data frame

```python
def idf_function(word):
    # assigning word repetitions of faced words
    try:
        word_oc = word_repetitions[word]+1 
    except:
        word_oc =1 # if code faces an error then it assigns to 1
    #buy using formula we will find an idf function's value
    return np.log(len(clear_list)/word_oc)
```

```python
def tf_idf(sentence):
    clear_sentence = re.split('[^a-z]', sentence) # getting all cleared sentenceses 
    #declaring some lists
    words_of_the_sentence = []
    tf_result=[] # empty list to assign here tf results 
    idf_result=[] # empty list to assign here idf results
    for word in clear_sentence: # condition to find out either word in our dictionary or not
        if word in final_dict:
            #tf
            tf_res = tf(sentence,word)
            tf_result.append(tf_res) # appending results of tf function 
            #idf
            idf_res = idf_function(word)
            idf_result.append(idf_res) # appending respectively received results in corresponding lists
            words_of_the_sentence.append(word)
    
    new_frame = pd.DataFrame(words_of_the_sentence) # creating dataframe to represent the result of functions
    new_frame['TF'] = tf_result
    new_frame['IDF'] = idf_result
    new_frame['TF*IDF'] = [a*b for a,b in zip(tf_result,idf_result)]
    new_frame.columns = ['Words', 'TF','IDF','TF*IDF']
    
    return new_frame
```
